```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the CSV file (pd.read_csv)
df = pd.read_csv('NY-House-Dataset.csv')
df_long = ['LONGITUDE']
df['price_per_sqft'] = df['PRICE'] / df['PROPERSTYQFT']

fig, ax = plt.subplots(figsize=(10, 6))

# plotting the columns
ax.scatter(
    df_long,
    df,
    color='steelBlue',
    s= 20,
    alpha= 0.6
)

ax.set_xlabel('Longitude')
ax.set_ylabel('Square feet')
ax.set_title('Square feet vs longitude')
plt.tight_layout()
plt.show()
```
