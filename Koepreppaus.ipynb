```python
class Dog:
    def __init__(self, name, breed, sound):
        self._name = name
        self._breed = breed
        self._sound = sound

    def describeSelf(self):
        print(f"My name is {self._name} and I am {self._breed}")

    def makesound(self):
        print(self._sound)

class Cat:
    def __init__(self, name, breed, sound):
        self._name = name
        self._breed = breed
        self._sound = sound

    def describeSelf(self):
        print(f"My name is {self._name} and I am a {self._breed}")

    def makeSound(self):
        print(self._sound)

Rex = Dog("Rex", "Labrador", "woof woof")
Rex.describeSelf()
Rex.makesound()

Patti = Cat("Patti", "Persian", "Meow")
Patti.describeSelf()
Patti.makeSound()
```

```python
class Animal:
    def __init__(self, name, race, sound):
        self._name = name
        self._race = race
        self._sound = sound

    def makeSound(self):
        print(self._sound)

    def describe(self):
        print(f"My name is {self._name} and I am {self._race}.")

class Dog(Animal):
    def __init__(self, name, race, sound):
        super().__init__(name, race, sound)

Rex = Dog("Rex", "Dog", "Woof woof")
Rex.describe()
Rex.makeSound()
```

```python
class Counter():
    def __init__(self):
        self._value = 0

    def Increment(self):
        self._value += 1
        return print("Added +1 to the count!")
    
    def getValue(self):
        return self._value
    
    def resetValue(self):
        self._value = 0
        return print("Counter value set to 0")

c = Counter()
c.Increment()
c.Increment()
c.getValue()
```

```python
class Vehicle():
    def __init__(self, make):
        self._make = make

    def describe(self):
        print(f"The vehicle is a {self._make}")

class Car(Vehicle):
    def __init__(self, make, doors):
        super().__init__(make)
        self._doors = doors

    def describe(self):
        print(f"The car is {self._make} and has {self._doors} doors")

class Motorcycle(Vehicle):
    def __init__(self, make, wheels):
        super().__init__(make)
        self._wheels = wheels

    def describe(self):
        print(f"Vehicle is {self._make} and has {self._wheels} wheels")

car = Car("Toyota", 4)
car.describe()

motorcycle = Motorcycle("Husqvarna Svartpilen 401", 2)
motorcycle.describe()
```
